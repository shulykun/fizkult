```python
import math
import itertools
import matplotlib.pyplot as plt

import numpy as np
from IPython.display import Audio
```

```python
np.pi
```

```python
fs = 10

time = 1
n = 6
angle = 6
grade = 1

step = 0.01

def polar_polygon(x,n, angle, grade):
    pp = 1 / math.cos((2/n)*math.asin(math.sin((n/2)*x + angle))) ** grade
    return pp * math.sin(x) 

samples_curve = np.array([])
for angle in range(30):
    i_gen = (polar_polygon(i,n,angle,grade) for i in itertools.count(start=0, step=step))
    
    samples_curve_ = np.array([next(i_gen )for i in range(int(np.pi * 2 / step))])

    samples_curve = list(samples_curve) + list(samples_curve_)

    samples_curve = np.array(samples_curve)
    
    
samples_curve = samples_curve / samples_curve.max() * 75
    
figure = plt.figure(figsize=(15, 20))
plt.subplot(3, 1, 1)
plt.axis('off')
plt.plot(samples_curve)
```
