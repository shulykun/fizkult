```python
# A general “Polar Polygon” function is:
# PPn(x) = sec((2/n)·arcsin(sin((n/2)·x)))
fs = 44100
limit = 1200
time = 30
N = 4

ANGLE1 = 1
x = 1
def polar_polygon(x,n, angle):
#     x = x + .005

    pp = 1 / math.cos((2/n)*math.asin(math.sin((n/2)*x+angle)))
    return pp * (math.sin((x % 4200)))

a = []
# for i in range(1, 100):
#     a.append(polar_polygon(i,n))
    
# i_gen = (polar_polygon(i,N,ANGLE1) for i in itertools.count(start=0, step=.035))

# samples = [next(i_gen ) for i in range(fs*time)]

# sns.lineplot(data=samples[:3000])

samples = []
for angle in range(0,2000,2):
    
    time = .1357 / 4
    
    i_gen = (polar_polygon(i,N,angle /10) for i in itertools.count(start=0, step=.0042))

    samples_chunk = [next(i_gen) for i in range(int(fs*time))]
    
    samples = samples + samples_chunk

offset = 700
samples_total = np.array(samples) + np.array(list(samples[offset:] + samples[-offset:] ))


figure = plt.figure(figsize=(15, 20))
plt.subplot(3, 1, 1)
plt.axis('off')
plt.plot(samples_total[:limit*80])


figure = plt.figure(figsize=(15, 20))
plt.subplot(3, 1, 1)
plt.axis('off')
plt.plot(samples_total[:limit*5])


Audio(samples_total, rate=fs)

# https://1ucasvb.tumblr.com/post/42906053623/in-a-previous-post-i-showed-how-to-geometrically
```
